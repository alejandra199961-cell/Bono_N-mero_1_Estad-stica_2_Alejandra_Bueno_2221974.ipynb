Manejo de módulos integrados

NumPy: se usa principalmente para trabajar con datos númericos de forma rápida, eficiente y estructurada. Un objeto tipo NumPy es un arreglo base para hacer matemáticas y manipular datos en Phyton.

Pandas: se usa para trabajar con datos estructurales (Como tablas de Excel o bases de datos). Un objeto tipo pandas permite cargar, limpiar, transformar y analizar datos. Lo usamos para bases de datos relacionales.

La diferencia entre NumPy y Pandas radica en su propósito y nivel de uso: NumPy es una liberia enfocada en el cálculo numérico eficiente mediante arrays multidimensionales, ideal para operaciones matemáticas rápidas, álgebra lineal y procesamiento de datos puramente numéricos; mientras que Pandas es una libreria de más alto nivel diseñada para la manipulación y análisis de datos estructurados, como tablas tipo Excel o bases de datos, permitiendo trabajar con columnas con nombres, datos de distintos tipos, limpieza, filtrado y agrupación; en la práctica, Pandas se apoya internamente de NumPy, por lo que ambos se complementan y se utilizan juntos en la mayoria de flujos de ciencia de datos.



# NumPy (Numerical Python)

Es la biblioteca fundamental para el cómputo científico en Python. Su fuerte es el manejo de arreglos multidimensionales (matrices) y el cálculo matemático de alta velocidad.

* **Uso en Ciencia de Datos:** Se utiliza en las capas más bajas de los algoritmos. Es la herramienta para realizar operaciones de álgebra lineal, transformaciones matriciales, generación de números aleatorios y operaciones matemáticas complejas sobre grandes volúmenes de datos numéricos.

### Ventajas para implementar algoritmos

* **Velocidad y eficiencia (Vectorización):** Los arreglos de NumPy (`ndarrays`) están escritos en C. Ejecutan operaciones en bloques de datos completos sin necesidad de usar bucles `for` en Python, lo que acelera masivamente el entrenamiento de algoritmos.
* **Eficiencia en memoria:** Almacena los datos en bloques contiguos de memoria y exige que todos los elementos sean del mismo tipo, lo que reduce drásticamente el espacio que ocupa en comparación con las listas nativas de Python.
* **Álgebra lineal nativa:** Viene con módulos específicos (`numpy.linalg`) indispensables para resolver sistemas de ecuaciones, calcular autovalores o descomponer matrices, la base de algoritmos como la Regresión Lineal o PCA (Análisis de Componentes Principales).

### Desventajas para implementar algoritmos

* **Rigidez de tipos (Homogeneidad):** Un arreglo de NumPy solo puede contener un tipo de datos a la vez (por ejemplo, solo flotantes). Si intentas mezclar texto y números, NumPy convertirá todo a cadenas, perdiendo la capacidad de hacer cálculos.
* **Estructura estática:** No maneja bien la inserción o eliminación de columnas o filas sobre la marcha; cambiar el tamaño de un arreglo suele implicar recrearlo por completo en memoria.
* **Falta de etiquetas:** Los datos se indexan estrictamente por su posición numérica (0, 1, 2...). No puedes nombrar las columnas, lo que dificulta trabajar con datos tabulares complejos.

# Pandas (Python Data Analysis Library)

Es la biblioteca estándar para la manipulación y el análisis de datos tabulares (estructurados en filas y columnas). Está construida directamente sobre NumPy.

* **Uso en Ciencia de Datos:** Domina las etapas de **limpieza, preparación y exploración de datos (EDA)**. Se usa para cargar archivos (CSV, Excel, SQL), manejar datos faltantes, filtrar, agrupar y transformar variables antes de pasarlas a un modelo de Machine Learning.

### Ventajas para implementar algoritmos

* **Estructuras heterogéneas (`DataFrames`):** Permite mezclar diferentes tipos de datos (columnas con texto, columnas con números, fechas, booleanos) en una sola tabla, reflejando fielmente las bases de datos de la vida real.
* **Indexación por etiquetas:** Puedes acceder a los datos usando los nombres reales de las variables o textos en las filas, haciendo que el código del algoritmo sea mucho más intuitivo y fácil de depurar.
* **Herramientas avanzadas de alineación y limpieza:** Facilita enormemente el manejo de valores nulos o faltantes (`NaN`), la fusión de tablas (`merge`, `join`) y la agrupación de datos (`groupby`), tareas que toman el 80% del tiempo antes de correr un algoritmo.

### Desventajas para implementar algoritmos

* **Alto consumo de memoria:** Pandas duplica o triplica fácilmente el tamaño del archivo original en la memoria RAM debido a los metadatos y los índices que genera para facilitar la búsqueda. No está optimizado para Big Data en una sola máquina.
* **Menor rendimiento en cálculo puro:** Aunque usa NumPy por debajo, añadir capas de abstracción (como las etiquetas de texto) hace que las operaciones matemáticas puras sean notablemente más lentas en Pandas que si se hicieran directamente en NumPy.
* **Sintaxis compleja y redundante:** Existen muchas formas de hacer una misma tarea (por ejemplo, filtrar filas), lo que puede dar lugar a errores sutiles como las advertencias de copia (`SettingWithCopyWarning`) si no se tiene cuidado al modificar datos dentro del algoritmo.
